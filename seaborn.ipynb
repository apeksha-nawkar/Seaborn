```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x=sns.load_dataset("tips")
x
```

```python
sns.scatterplot(x='day',y='tip', data=x)
```

```python
 g=sns.FacetGrid(x, col="tip", row="sex" ,size=2.4)
g.map(sns.scatterplot,"total_bill", "tip")
```

```python
sns.boxplot(x='day',y='tip', data=x)
```

```python
fig,(ax1,ax2,ax3,ax4, ax5, ax6, ax7)=plt.subplots(1,7, figsize=(46,10))
sns.boxplot(x='day',y='tip', data=x, ax=ax1)
sns.barplot(x='day',y='tip', data=x, ax=ax2)
sns.stripplot(x='day',y='tip', data=x, ax=ax3, hue='sex', jitter=True)
sns.scatterplot(x='day',y='tip', data=x,  hue='sex',ax=ax4)
sns.swarmplot(x='day',y='tip', data=x, ax=ax5)
sns.histplot(x='day',y='tip', hue='sex', data=x, ax=ax6)
sns.violinplot(x="day", y="tip", data=x)
```

```python
sns.violinplot(x="day", y="tip", data=x, inner=None, color="lightpink")
sns.swarmplot(x='day',y='tip', data=x)
```

```python
sns.jointplot(x="total_bill", y="smoker",data=x)
```

```python
sns.pairplot(x, hue="sex")
```
